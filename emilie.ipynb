```python
import pandas as pd
from random import randint
```

## Create fake dataset of schools and scores

```python
def get_dataframe():
    """
    Creates the Fake DataFrame
    """
    
    a = {'school': ['LSU', 'Alabama', 'CSUF', 'JHU', 'Oregon'],
         'sports': [randint(1,10) for x in range(5)], 'science': [randint(1,10) for x in range(5)],
         'humanities':[randint(1,10) for x in range(5)], 'debate': [randint(1,10) for x in range(5)],
         'technology':[randint(1,10) for x in range(5)], 'research': [randint(1,10) for x in range(5)],
         'affordability': [randint(1,10) for x in range(5)]
        }

    df = pd.DataFrame(a)
    return df
```

```python
get_dataframe()
```

## Create function that creates a new DataFrame sorted by rank

```python
def factors(df, factor_list, weights, ascending=False):
    """
    Rank schools based on factors
    
    Params
    ------
    df : pd.DataFrame
         DataFrame containing school and scores
    factor_list : list
         Contains the factors you want to rank by 
    weights : list
         contains weights
         
    Returns
    -------
    final_ranking : pd.DataFrame
         A DataFrame with the schools ranked by the mean
         score for the factors you selected
         
    Notes
    -----
    You can access the school names by looking at the index
    """
    
    df['rank'] = (df[factor_list]*weights).mean(axis=1)
    df.set_index('school', inplace=True)
    final_ranking = df.sort_values('rank', ascending=ascending)
    return final_ranking
```

```python
weights = [0.6, 0.3, 0.1]
factor_list = ['sports', 'debate', 'research']

df = get_dataframe()
factors(df, factor_list, weights, ascending=False)
```
